```
#자동차 클래스 (상속을 위한 설계....)

class Car : 
  speed = 0
  
  def upSpeed(self, value):
      self.speed += value
      print("현재 속도(슈퍼 클래스) : ",self.speed)

xxx = Car()
xxx.upSpeed(80)
xxx.upSpeed(100)

class Sedan(Car): #Car를 상속 받음
    def upSpeed(self,value): #오버라이딩
      self.speed +=value
      if(self.speed >= 150): #세단의 경우 속도가 150이 넘어가면 제한
        self.speed = 150
      print("현재 속도(세단 클래스) : ",self.speed)

class Truck(Car): #Car를 상속 받음
  pass


class Sonata(Sedan): #Sedan을 상속 받음 (Car -> Sedan -> Sonata)
  pass

ss = Sedan()
ss.upSpeed(70)
ss.upSpeed(70)
ss.upSpeed(70)
ss.upSpeed(70)

tt = Truck()
tt.upSpeed(70)
tt.upSpeed(70)
tt.upSpeed(70)
tt.upSpeed(70)

sn = Sonata()
sn.upSpeed(90)
sn.upSpeed(90)
sn.upSpeed(90) #세단의 method인 upSpeed 호출
```
