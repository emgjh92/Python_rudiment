```
print("+++++++++++++++++딕셔너리+++++++++++++++++++++++")
#값의 중복은 허용하지만, Key의 중복은 허용하지 않는다.
#파이썬의 딕셔너리 = JAVA의 Map = JavsScript 의 Object 
#(JavaScript는 Object 와 Map 의 경계가 없다...)
dic1 = {"a":1,"b":2,"c":3} #일반적으론 앞에 키는 String, 뒤에껀 값

print(dic1) #출력값 : {'a': 1, 'b': 2, 'c': 3}
print(dic1["a"]) #출력값 : 1

dic1 = {}
print(dic1)
dic1["a"]=10
dic1["b"]=20
dic1["c"]=30

print(dic1) #출력값 : {'a': 10, 'b': 20, 'c': 30}
print(len(dic1)) #출력값 : 3


dic1 = {}
dic1["a class"] = {40,30,20,10}
dic1["b class"] = {40,30,20,10}
dic1["c class"] = {40,30,20,10}

print(dic1)
#출력값 : {'a class': {40, 10, 20, 30}, 'b class': {40, 10, 20, 30}, 'c class': {40, 10, 20, 30}}

dic2 = {"a" : [1,2,3],"b" : [3,3,3], "c" : 70}
print(dic2) #출력값 : {'a': [1, 2, 3], 'b': [3, 3, 3], 'c': 70}
print(dic2["a"][1]) #출력값 : 2

for data in dic2["a"] :
  print(data)
  #출력값 : 1 2 3


list2 = [{"a" : 1, "b":2}, {"c" : 1, "d":2}] #리스트 안에 딕셔너리 형태
print(list2) #출력값 : [{'a': 1, 'b': 2}, {'c': 1, 'd': 2}]

print(list2[0]["b"]) #출력값 : 2


aa = {"a" : 1, "b" : 2}
print(aa.get("a")) #출력값 : 1
print(aa["a"]) #출력값 : 1


print(aa.get("c")) #출력값 : None (예외는 아님)
#print(aa["c"]) #출력값 : ERROR (예외발생)

aa.keys() #출력값 : dict_keys(['a', 'b'])

type(aa.keys()) #출력값 : dict_keys

keys = list(aa.keys()) # dict_keys 를 '진짜'리스트로 변환

print(aa.values())   #dict_values([1, 2])

xxx = "b" in aa # "b"라는 키가 딕셔너리 aa 안에 존재하면 true 아니면 false 반환
print(xxx)
```

```
print("+++++++++++++++++컴프리헤션+++++++++++++++++++++++")
#파이썬에만 존재하는 문법

numList = []

for x in range(1,6,1):
  numList.append(x)

print(numList) #출력값 : [1, 2, 3, 4, 5]

################위 코드를 컴프리헤션으로 바꾸면#################

numList = [num for num in range(1,6,1)]

print(numList) #출력값 : [1, 2, 3, 4, 5]


numList = [num*num for num in range(1,6,1)]

print(numList) #출력값 : [1, 4, 9, 16, 25]


numList = [num*num for num in range(1,6,1) if num%2==0] 
#num이 짝수인 경우에만 제곱해서 append 하겠다.

print(numList) #출력값 : [4, 16] 
```

```
print("+++++++++++++++++zip 함수+++++++++++++++++++++++")

foods = ['a','b','c','d','e']
sides = ['1','2','3','4','5']

#다른 언어와 다르게 파이썬은 함수의 return 값이 여러개가 될 수 있다.

for f,s in zip(foods,sides):
  print(f,s)

  # a 1
  # b 2
  # c 3
  # d 4
  # e 5
print("----------------------------------------------")
  #그럼 서로 갯수가 다를땐?
foods = ['a','b','c','d','e']
sides = ['1','2','3']

for f,s in zip(foods,sides):
  print(f,s)

  # a 1
  # b 2
  # c 3
#세개만 출력된다. (foods 가 숫자가 적을 경우엔 foods 갯수 기준으로 간다)



xxx = list(zip(foods,sides)) #리스트로 변환
print(xxx)
xxx = dict(zip(foods,sides)) #딕셔너리로 변환
print(xxx)
```

```
print("+++++++++++++++++리스트의 복사+++++++++++++++++++++++")

aa = [11,22,33,44,55,66,77,88]
bb = aa #얕은복사 = 같은 참조주소를 가진다.
print(bb) #출력값 : [11, 22, 33, 44, 55, 66, 77, 88]
#a 와 b가 동일한 메모리 주소값을 가르키게 만든다.

bb[3] = 1000
print(aa) #[11, 22, 33, 1000, 55, 66, 77, 88]
print(bb) #[11, 22, 33, 1000, 55, 66, 77, 88]
#bb값과 함께 aa도 바뀌어 버렸다.

print("----------------------------------------------")
aa = [11,22,33,44,55,66,77,88]
bb=aa[:] #깊은 복사 (aa의 전체를 슬라이싱 해서 bb에 넣어준다.)
#서로 다른 메모리 주소 값을 가진다.
bb[3] = 1000
print(aa)#[11, 22, 33, 44, 55, 66, 77, 88]
print(bb)#[11, 22, 33, 1000, 55, 66, 77, 88]



```
